# 1. Загрузка библиотек и написание вспомогательных функций

```python
from package.libraries import *
from package.statistic_functions import *
from package.preparation_functions import *
```

```python
## Подготовка вспомогательных функций, используемых в исследовании
def pvalue_info(pvalue):
    """
        Функция вывода информации о параметре pvalue, полученного в результате статистического теста

            Параметры:
                pvalue (float): значение pvalue
            Выходные параметры (None)

    """

    print(f'pvalue = {(pvalue * 100):.6f}%')
    if pvalue > 0.05:
        print('Выбираем гипотезу H0')
    else:
        print('Выбираем гипотезу H1')


def paired_T_test(df_1, df_2, check):
    """
        Функция запуска теста по парному Т-критерию

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
            Выходные параметры (None)

    """
    if pd.isna(check):
        pvalue = stats.ttest_rel(df_1, df_2)[1]
        print('Выбранный тест: парный Т-критерий\n')
        pvalue_info(pvalue)
    else:
        pvalue = stats.ttest_rel(df_1, df_2, alternative='less')[1]
    return pvalue


def student_T_test(df_1, df_2, check):
    """
        Функция запуска теста по Т-критерию Стьюдента

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
            Выходные параметры (None)

    """
    if pd.isna(check):
        pvalue = stats.ttest_ind(df_1, df_2)[1]
        print('Выбранный тест: Т-критерий Стьюдента\n')
        pvalue_info(pvalue)
    else:
        pvalue = stats.ttest_ind(df_1, df_2, alternative='less')[1]
    return pvalue


def welch_T_test(df_1, df_2, check):
    """
        Функция запуска теста по Т-критерию Уэлча

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
            Выходные параметры (None)

    """
    if pd.isna(check):
        pvalue = stats.ttest_ind(df_1, df_2)[1]
        print('Выбранный тест: Т-критерий Уэлча\n')
        pvalue_info(pvalue)
    else:
        pvalue = stats.ttest_ind(df_1, df_2, alternative='less')[1]
    return pvalue


def wilcoxon_RANK_test():
    """
        Функция запуска теста по ранговому критерию Уилкоксона

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
            Выходные параметры (None)

    """
    if pd.isna(check):
        pvalue = stats.wilcoxon(df_1, df_2)[1]
        print('Выбранный тест: знаковый ранговый критерий Уилкоксона\n')
        pvalue_info(pvalue)
    else:
        pvalue = stats.wilcoxon(df_1, df_2, alternative='less')[1]
    return pvalue
    

def mannwhitney_U_test(df_1, df_2, check):
    """
        Функция запуска теста по критерию Манна-Уитни

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
            Выходные параметры (None)

    """
    if pd.isna(check):
        pvalue = stats.mannwhitneyu(df_1, df_2)[1]
        print('Выбранный тест: критерий Манна-Уитни\n')
        pvalue_info(pvalue)
    else:
        pvalue = stats.mannwhitneyu(df_1, df_2, alternative='less')[1]
    return pvalue


# Функция выбора статистического теста для проверки гипотезы
def shapiro_test(df_1, df_2, check):
    """
        Функция проверки двух выборок на распределение по нормальному закону

            Параметры:
                df_1 (DataFrame): первая проверяемая выборка
                df_2 (DataFrame): вторая проверяемая выборка
            Выходные параметры (bool)

    """
    s_df_1 = round(stats.shapiro(df_1)[1] * 100, 6)
    s_df_2 = round(stats.shapiro(df_2)[1] * 100, 6)
    if pd.isna(check):
        print(' ЗНАЧЕНИЯ pvalue ВЫБОРОК')
        print('  на тесте Шапира-Уилка')
        print('============================')
        print(f' Первая выборка: {s_df_1:.6f}%')
        print(f' Вторая выборка: {s_df_2:.6f}%')
        print('----------------------------')
    if s_df_1 <= .05 or s_df_2 <= 0.05:
        if pd.isna(check):
            print('Выборки НЕПАРАМЕТРИЧЕСКИЕ (не распеределены по нормальному закону распределения)')
        return False
    else:
        if pd.isna(check):
            print('Выборки ПАРАМЕТРИЧЕСКИЕ (распеределены по нормальному закону распределения)')
        return True


def choose_method(df_1, df_2, dependenceFlag=True, check=None):
    """
        Функция выбора теста для проверки гипотез

            Параметры:
                df_1 (DataFrame): первая сравниваемая выборка
                df_2 (DataFrame): вторая сравниваемая выборка
                dependenceFlag (bool): флаг зависимости двух выборок
            Выходные параметры (None)

    """
    # Проверяем распределения выборок по нормальному закону
    normalFlag = shapiro_test(df_1, df_2, check)
    # Указываем на зависимость или независимость выборок
    if pd.isna(check):
        if dependenceFlag:
            print('Выборки ЗАВИСИМЫЕ (парам в 1 выборке соответствуют пары во 2 выборке)\n')
        else:
            print('Выборки НЕ ЗАВИСИМЫЕ (парам в 1 выборке не соответствуют пары во 2 выборке)\n')

    # Выбор теста для проверки гипотезы
    if normalFlag and dependenceFlag:
        paired_T_test(df_1, df_2, check)
    elif normalFlag and not dependenceFlag:
        if stats.levene(df_1, df_2)[1] <= 0.05:
            if pd.isna(check):
                print('ДИСПЕРСИИ выборок - неравны')
            welch_T_test(df_1, df_2, check)
        else:
            if pd.isna(check):
                print('ДИСПЕРСИИ выборок - равны')
            student_T_test(df_1, df_2, check)
    elif not normalFlag and dependenceFlag:
        wilcoxon_RANK_test(df_1, df_2, check)
    else:
        return mannwhitney_U_test(df_1, df_2, check)

## все разработанные процедуры перенесены в модуль functions.py директории package
```

# 6. Ответы на вопросы, заданные в п.2.4 пп.4-6

## 6.1. Из каких источников / кампаний / устройств / локаций к нам идёт самый целевой трафик (и с точки зрения объёма трафика, и с точки зрения CR)

### 6.1.1. Выбираем каналы с наибольшим целевым трафиком

```python
# Агрегируем событий датафрейм по источникам по дням с указанием целевого события
min_date, max_date = dt.date(2222, 12, 31),  dt.date(1900, 1, 1)
max_period = 0
agg_function = {'session_id': 'count', 'conversion_rate': 'sum'}
utm_source_traffic = dict()
for elem in df_sessions.utm_source.unique():
    df_sessions_source = df_sessions.\
                        loc[df_sessions.utm_source == elem].\
                        groupby('visit_date').\
                        agg(agg_function)
    df_sessions_source['CR'] = round((df_sessions_source['conversion_rate'] /\
                               df_sessions_source['session_id']) * 100, 2)
    df_sessions_source.drop(['session_id'], axis=1, inplace=True)
    utm_source_traffic[elem] = df_sessions_source
    period = len(df_sessions_source)
    if utm_source_traffic[elem].index.min() < min_date:
        min_date = utm_source_traffic[elem].index.min()
    if utm_source_traffic[elem].index.max() > max_date:
        max_date = utm_source_traffic[elem].index.max()
    if period > max_period:
        period = max_period
```

```python
# Приводим датафреймы трафика к единому размеру
# (добавляем нулевые значения трафика в даты когда трафик отсутствовал)
all_index = pd.date_range(min_date, max_date).date
for elem in utm_source_traffic:
    if len(elem) < len(all_index):
        non_data_index = set(all_index) - set(utm_source_traffic[elem].index)
        value = [0.0 for _ in range(len(non_data_index))]
        df_non_index = pd.DataFrame({'date': list(non_data_index), 'conversion_rate': value, 'CR': value})
        df_non_index = df_non_index.set_index(['date'])
        utm_source_traffic[elem] = pd.concat([utm_source_traffic[elem], df_non_index])
```

```python
# Формируем список трафиков для сортировки по наибольшей статистической величине
# (для статистического сравнения оставляем только датафреймы с ненулевым трафиком)
dst = [(name, df) for name, df in utm_source_traffic.items() if df.CR.sum() != 0]
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины CR
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].CR, \
                         dst[j+1][1].CR,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j], dst[j+1] =\
            dst[j+1], dst[j]
```

```python
# Отобразим первые 3 канала с наибольшим целевым трафиком по CR
fig, axs = plt.subplots(figsize=(15, 6))
axs.set_title('Каналы с наибольшим целевым трафиком по CR (%)')
axs.plot(dst[-1][1].sort_index().index, dst[-1][1].CR, color='green', label=dst[-1][0])
axs.plot(dst[-2][1].sort_index().index, dst[-2][1].CR, color='blue', label=dst[-2][0])
axs.plot(dst[-3][1].sort_index().index, dst[-3][1].CR, color='yellow', label=dst[-3][0])

axs.legend()
plt.show();
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины объёма трафика
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].conversion_rate, \
                         dst[j+1][1].conversion_rate,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j][1].conversion_rate, dst[j+1][1].conversion_rate =\
            dst[j+1][1].conversion_rate, dst[j][1].conversion_rate
```

```python
# Отобразим первые 3 канала с наибольшим целевым трафиком по объёму
fig, axs = plt.subplots(figsize=(15, 6))
axs.set_title('Каналы с наибольшим целевым трафиком по объёму (ед.визитов)')
axs.plot(dst[-1][1].sort_index().index, dst[-1][1].conversion_rate, color='green', label=dst[-1][0])
axs.plot(dst[-2][1].sort_index().index, dst[-2][1].conversion_rate, color='blue', label=dst[-2][0])
axs.plot(dst[-3][1].sort_index().index, dst[-3][1].conversion_rate, color='yellow', label=dst[-3][0])

axs.legend()
plt.show();
```
