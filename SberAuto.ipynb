# 4. Формирование датасета визитов с признаком целевого события

```python
# Формируем список полей-признака целевого события в датасете событий
target_sign = ['sub_car_claim_click', 'sub_car_claim_submit_click',
               'sub_open_dialog_click', 'sub_custom_question_submit_click',
               'sub_call_number_click', 'sub_callback_submit_click',
               'sub_submit_success', 'sub_car_request_submit_click']
```

```python
# Формируем вспомогательное поле для агрегации событий по визитам
df_hits['session'] = df_hits['session_id']
```

```python
# Формируем поле с признаком целевого события, совершённого клиентом
df_hits['conversion_rate'] = df_hits.event_action.\
                                apply(lambda x: 1 if x in target_sign else 0)
```

```python
# Агрегируем датафрейм событий по визитам с указанием наличия целевого события
agg_function = {'session_id': 'max', 'conversion_rate': 'max'}
df_hits_ag = df_hits.groupby('session').agg(agg_function)
```

```python
# Добавляем в датафрейм визитов признак целевого события
df_sessions = df_sessions.merge(df_hits_ag, how='left', on='session_id')
```

```python
# Добавляем в датасет визитов признак платного трафика
df_sessions['paid_traffic'] = df_sessions.utm_medium.\
apply(lambda x: 0 if x in ['organic', 'referral', '(none)'] else 1)
```

```python
# Делим датасет на датасеты платного и органического трафиков
df_sessions_paid = df_sessions.loc[df_sessions['paid_traffic'] == 1]
df_sessions_organic = df_sessions.loc[df_sessions['paid_traffic'] == 0]
```

```python
# Проводим агрегацию датасетов визитов с целевыми событиями для определения подневного трафика
agg_function = {'session_id': 'count', 'conversion_rate': 'sum'}
## на платном трафике
df_paid_CR = df_sessions_paid.groupby('visit_date').agg(agg_function)
## на органическом трафике
df_organic_CR = df_sessions_organic.groupby('visit_date').agg(agg_function)
```

```python
# Расчитываем CR для платного и органического трафиков
df_paid_CR['CR'] = round(100 * df_paid_CR['conversion_rate']\
                         / df_paid_CR['session_id'], 2)
df_organic_CR['CR'] = round(100 * df_organic_CR['conversion_rate']\
                         / df_organic_CR['session_id'], 2)
```

```python
# Визуализируем распределение полученного органического и платного трафика
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 2, figsize=(12,6))
axs[0].hist(df_organic_CR['CR'], bins=120, color='green')
axs[0].set_title('CR по органическому трафику')
axs[1].hist(df_paid_CR['CR'], bins=120, color='black')
axs[1].set_title('CR по платному трафику')
plt.show()
```

Нулевая и альтернативная гепотизы:  
 - $H_0$ - Conversion Rate (далее - CR) органического трафика не отличается от CR платного.  
 - $H_1$ - CR органического трафика отличается от CR платного трафика.  
 
 Выборки независимые, так как 73% клиентов, заходивших на сайт, относятся только к органическому или только к платному трафику.

```python
# Выбираем тест и проверяем гипотизу
choose_method(df_organic_CR['CR'], df_paid_CR['CR'], dependenceFlag=False)
```

```python
# Дополнительно проверим - больше ли платный трафик чем органический
stats.mannwhitneyu(df_paid_CR['CR'], df_organic_CR['CR'], alternative='greater')
```

**Вывод: Органический трафик отличается от платного с точки зрения CR.**  
Кроме того, дополнительно выяснено что платный трафик с точки зрения CR больше чем органический трафик.

## 6.1. Из каких источников / кампаний / устройств / локаций к нам идёт самый целевой трафик (и с точки зрения объёма трафика, и с точки зрения CR)

### 6.1.1. Определение источников с наибольшим целевым трафиком

```python
# Агрегируем датафрейм событий по источникам по дням с указанием целевого события
min_date, max_date = dt.date(2222, 12, 31),  dt.date(1900, 1, 1)
max_period = 0
agg_function = {'session_id': 'count', 'conversion_rate': 'sum'}
utm_source_traffic = dict()
for elem in df_sessions.utm_source.unique():
    df_sessions_source = df_sessions.\
                        loc[df_sessions.utm_source == elem].\
                        groupby('visit_date').\
                        agg(agg_function)
    df_sessions_source['CR'] = round((df_sessions_source['conversion_rate'] /\
                               df_sessions_source['session_id']) * 100, 2)
    df_sessions_source.drop(['session_id'], axis=1, inplace=True)
    utm_source_traffic[elem] = df_sessions_source
    period = len(df_sessions_source)
    if utm_source_traffic[elem].index.min() < min_date:
        min_date = utm_source_traffic[elem].index.min()
    if utm_source_traffic[elem].index.max() > max_date:
        max_date = utm_source_traffic[elem].index.max()
    if period > max_period:
        period = max_period
```

```python
# Приводим датафреймы трафика к единому размеру
# (добавляем нулевые значения трафика в даты когда трафик отсутствовал)
all_index = pd.date_range(min_date, max_date).date
for elem in utm_source_traffic:
    if len(elem) < len(all_index):
        non_data_index = set(all_index) - set(utm_source_traffic[elem].index)
        value = [0.0 for _ in range(len(non_data_index))]
        df_non_index = pd.DataFrame({'date': list(non_data_index), 'conversion_rate': value, 'CR': value})
        df_non_index = df_non_index.set_index(['date'])
        utm_source_traffic[elem] = pd.concat([utm_source_traffic[elem], df_non_index])
```

```python
# Формируем список трафиков для сортировки по наибольшей статистической величине
# (для статистического сравнения оставляем только датафреймы с ненулевым трафиком)
dst = [(name, df) for name, df in utm_source_traffic.items() if df.CR.sum() != 0]
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины CR
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].CR, \
                         dst[j+1][1].CR,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j], dst[j+1] =\
            dst[j+1], dst[j]
```

```python
# Отображаем первые 3 канала с наибольшим целевым трафиком по CR
fig, axs = plt.subplots(figsize=(15, 6))
axs.set_title('Каналы с наибольшим целевым трафиком по CR (%)')
axs.plot(dst[-1][1].sort_index().index, dst[-1][1].CR, color='green', label=dst[-1][0])
axs.plot(dst[-2][1].sort_index().index, dst[-2][1].CR, color='blue', label=dst[-2][0])
axs.plot(dst[-3][1].sort_index().index, dst[-3][1].CR, color='yellow', label=dst[-3][0])

axs.legend()
plt.show();
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины объёма трафика
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].conversion_rate, \
                         dst[j+1][1].conversion_rate,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j], dst[j+1] =\
            dst[j+1], dst[j]
```

```python
# Отображаем первые 3 источника с наибольшим целевым трафиком по объёму
fig, axs = plt.subplots(figsize=(15, 6))
axs.set_title('Каналы с наибольшим целевым трафиком по объёму (ед.визитов)')
axs.plot(dst[-1][1].sort_index().index, dst[-1][1].conversion_rate, color='green', label=dst[-1][0])
axs.plot(dst[-2][1].sort_index().index, dst[-2][1].conversion_rate, color='blue', label=dst[-2][0])
axs.plot(dst[-3][1].sort_index().index, dst[-3][1].conversion_rate, color='yellow', label=dst[-3][0])

axs.legend()
plt.show();
```

```python
# Приводим датафреймы трафика к единому размеру
# (добавляем нулевые значения трафика в даты когда трафик отсутствовал)
all_index = pd.date_range(min_date, max_date).date
for elem in geo_city_traffic:
    if len(elem) < len(all_index):
        non_data_index = set(all_index) - set(geo_city_traffic[elem].index)
        value = [0.0 for _ in range(len(non_data_index))]
        df_non_index = pd.DataFrame({'date': list(non_data_index), 'conversion_rate': value, 'CR': value})
        df_non_index = df_non_index.set_index(['date'])
        geo_city_traffic[elem] = pd.concat([geo_city_traffic[elem], df_non_index])
```

```python
# Формируем список трафиков для сортировки по наибольшей статистической величине
# (для статистического сравнения оставляем только датафреймы с ненулевым трафиком)
dst = [(name, df) for name, df in geo_city_traffic.items() if df.CR.sum() != 0]
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины CR
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].CR, \
                         dst[j+1][1].CR,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j], dst[j+1] =\
            dst[j+1], dst[j]
```

Поскольку за один визит клиент мог спросить о нескольких моделях автомобилей, то анализ спроса целесооразно проводить на датасете событий, в котором зарегистрированы все действия клиента сайте.

```python
# Добавим в датасет событий колонку с маркой и моделью автомобиля
# (в тех случаях если клиент просматривал конкретную марку и модель,
# выбирая их в каталоге 'sberauto.com/cars/all/')
df_hits['car'] = df.hit_page_path.apply(lambda x:
                                             '/'.join(x.split('/')[3:5])
                                             if len(str(x)) > 23 and
                                                 x[:22] == 'sberauto.com/cars/all/'
                                             else None)
df_hits.shape
```

```python
# Удаляем строки с неопределёнными маркой и моделью автомобиля
df = df_hits.dropna(subset=['car'])
df.shape
```

```python
# Сформируем датасет с уникальными просмотрами конкретного автомобиля
# в рамках одного визита
agg_function = {'session_id': 'max', 'car': 'max', 'hit_date': 'max', 'conversion_rate': 'max'}
df_uniquecar_visit = df.groupby(['session_id', 'car']).agg(agg_function)
df_uniquecar_visit.shape
```

```python
# Преобразуем атрибут даты в дату в формат datetime
df_uniquecar_visit.hit_date = df_uniquecar_visit.hit_date.apply(lambda x:
                                    dt.datetime.strptime(x, '%Y-%m-%d').date())
```
